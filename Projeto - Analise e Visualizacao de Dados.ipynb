```python
# Resposta questão 1: 
total = df['Total'].count() #contando a partir da coluna 'total'
print("Reclamações Registradas: ", total)
```

```python
# Resposta questão 2: 
tempo_total = df['Tempo Resposta'].sum()
media_tempo = tempo_total / total
t_max = df['Tempo Resposta'].max()
t_min = df['Tempo Resposta'].min()
print(f"Tempo médio de resposta: {media_tempo: .1f}") #arredondado para uma casa decimal
print("Tempo máximo de resposta: ", t_max)
print("Tempo mínimo de resposta: ", t_min)
```

```python
# Resposta questão 3: 
nota_soma = df['Nota do Consumidor'].sum()
media_nota = nota_soma / total
n_max = df['Nota do Consumidor'].max()
n_min = df['Nota do Consumidor'].min()
print(f"Nota média: {media_nota: .1f}") #arredondado para uma casa decimal
print("Nota máxima: ", n_max)
print("Nota mínima: ", n_min)
```

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Resposta questão 5a: 
#reclamações por sexo
r_sexo = df['Sexo'].value_counts() 
print("Reclamações Registradas por sexo M e F: ")
print(r_sexo)

#gerando grafico de frequencia por sexo
x = ['Masculino', 'Feminino']
y = [r_sexo[0], r_sexo[1]]


barlist=plt.bar(x,y)
barlist[0].set_color('aqua')
barlist[1].set_color('lime')
plt.xlabel('SEXO')
plt.ylabel('QTD RECLAMAÇÕES')
plt.title('Reclamações por Sexo')

plt.show()
```

```python
# Resposta questão 5b: 
r_estado = df['UF'].value_counts()
print("Reclamações Registradas Estado(UF): ")
print(r_estado)

plt.rcParams['figure.figsize'] = (12,12) #aumentando o tamanho para ficar mais vizivel 
#criando o grafico
plt.bar(df['UF'].value_counts().keys(), df['UF'].value_counts())
plt.xlabel('Estados')
plt.ylabel('QTD Reclamações')
plt.title('Reclamações por Estado')
plt.show()



```

```python
# Resposta questão 5c: 
resp = df['Respondida'].value_counts()

print("Número de reclamações não respondidas: ", resp['N'])
print("Número de reclamações respondidas: ", resp['S'])

barlist2 = plt.bar(df['Respondida'].value_counts().keys(), df['Respondida'].value_counts())
barlist2[1].set_color('red')
plt.xlabel('RESPOSTAS')
plt.ylabel('QTD RESPONDIDAS')
plt.title('Reclamações Respondidas')
plt.show()
```

```python
# Resposta questão 6a: 
barlist3 = plt.bar(df['Avaliação Reclamação'].value_counts().keys(), df['Avaliação Reclamação'].value_counts())
barlist3[0].set_color('green')
barlist3[1].set_color('aqua')
barlist3[0].set_color('red')
plt.xlabel('SITUAÇÃO')
plt.ylabel('QTD AVALIAÇÕES')
plt.title('Gráfico das Situações dos Casos')
plt.show()
```

```python
# Resposta questão 6b: 
barlist4 = plt.bar(df['Faixa Etária'].value_counts().keys(), df['Faixa Etária'].value_counts())

plt.tick_params(axis='x', which = 'major',labelsize = 8.5)
my_colors = ['brown','pink', 'red', 'green', 'blue', 'cyan','orange','purple']
for k in range(len(barlist4)):
    barlist4[k].set_color(my_colors[k])

   
plt.xlabel('FAIXA ETÁRIA')
plt.ylabel('QTD POR FAIXA')
plt.title('Gráfico da Faixa etária das reclamações')
plt.show()
```
